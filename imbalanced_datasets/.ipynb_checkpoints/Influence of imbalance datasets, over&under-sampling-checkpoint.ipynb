# Impact of Imbalanced Datasets on Machine Learning Algorithms

## Logistic Regression on Imbalanced Dataset

```python
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
import numpy as np
import matplotlib.pyplot as plt

def scatter_plot(X, y, plot_boundary=False, b=None, w1=None, w2=None, title=None):
    x1_min, x1_max = X[:,0].min()-1, X[:,0].max()+1
    x2_min, x2_max = X[:,1].min()-1, X[:,1].max()+1
    plt.scatter(X[y==0, 0], X[y==0, 1], s=10, alpha=0.5, label='Negative')
    plt.scatter(X[y==1, 0], X[y==1, 1], s=10, alpha=0.5, label='Positive')
    if plot_boundary:
        intercept = -b/w2 # in x1,x2 plate
        slope = -w1/w2
        x1 = np.array([x1_min, x1_max])
        x2 = slope*x1 + intercept
        plt.plot(x1, x2, 'k', lw=1, ls='--')
        plt.fill_between(x1, x2, x2_min, color='tab:orange', alpha=0.2)
        plt.fill_between(x1, x2, x2_max, color='tab:blue', alpha=0.2)
    plt.xlim(x1_min, x1_max)
    plt.ylim(x2_min, x2_max)
    plt.xlabel('x1')
    plt.ylabel('x2')
    plt.legend()
    plt.show()
```

```python
X, y = make_classification(
    n_samples=100000, 
    n_features=2, 
    n_redundant=0,
    n_clusters_per_class=1, 
    weights=[0.999], 
    flip_y=0, 
    random_state=110
)
scatter_plot(X, y)
```

```python
clf = LogisticRegression()
clf.fit(X, y)
b = clf.intercept_[0]
w1, w2 = clf.coef_.T

# decision boundary
scatter_plot(X, y, plot_boundary=True, b=b, w1=w1, w2=w2)
```

## Decision Tree on Imbalanced Dataset

```python
from sklearn.inspection import DecisionBoundaryDisplay
from sklearn.tree import DecisionTreeClassifier

def plot_dt_boundary(clf, X, y):
    disp = DecisionBoundaryDisplay.from_estimator(clf, X, alpha=0.3, xlabel='x1', ylabel='x2', response_method='predict', cmap='RdBu_r')
    disp.ax_.scatter(X[y==0, 0], X[y==0, 1], s=10, alpha=0.5, label='Negative')
    disp.ax_.scatter(X[y==1, 0], X[y==1, 1], s=10, alpha=0.5, label='Positive')
    disp.ax_.legend()
    plt.show()
```

```python
X, y = make_classification(
    n_samples=1000, 
    n_features=2, 
    n_redundant=0,
    n_clusters_per_class=1, 
    weights=[0.90], 
    flip_y=0, 
    random_state=37,
    class_sep=0.3,
    scale=0.6
)

clf = DecisionTreeClassifier(criterion='entropy', max_depth=5)
clf.fit(X, y)
plot_dt_boundary(clf, X, y)
```

```python
X, y = make_classification(
    n_samples=1000, 
    n_features=2, 
    n_redundant=0,
    n_clusters_per_class=2, 
    weights=[0.90], 
    flip_y=0, 
    random_state=1,
)

clf = DecisionTreeClassifier(criterion='entropy', max_depth=5)
clf.fit(X, y)
plot_dt_boundary(clf, X, y)
```

# Over-sampling & Under-sampling

## Over-sampling

```python
X, y = make_classification(
    n_samples=100000, 
    n_features=2, 
    n_redundant=0,
    n_clusters_per_class=1, 
    weights=[0.999], 
    flip_y=0, 
    random_state=110
)
scatter_plot(X, y)
```

```python
from imblearn.over_sampling import RandomOverSampler

ros = RandomOverSampler(random_state=42, sampling_strategy='minority')
X_ros, y_ros = ros.fit_resample(X, y)

scatter_plot(X_ros, y_ros)
```

```python
clf = LogisticRegression()
clf.fit(X_ros, y_ros)
b_ros = clf.intercept_[0]
w1_ros, w2_ros = clf.coef_.T

# new decision boundary on over-sampled dataset
scatter_plot(X_ros, y_ros, plot_boundary=True, b=b_ros, w1=w1_ros, w2=w2_ros)
```

## Under-sampling

```python
from imblearn.under_sampling import RandomUnderSampler

rus = RandomUnderSampler(random_state=42, sampling_strategy='majority')
X_rus, y_rus = rus.fit_resample(X, y)

scatter_plot(X_rus, y_rus)
```

```python
clf = LogisticRegression()
clf.fit(X_rus, y_rus)
b_rus = clf.intercept_[0]
w1_rus, w2_rus = clf.coef_.T

# new decision boundary on under-sampled dataset
scatter_plot(X_rus, y_rus, plot_boundary=True, b=b_rus, w1=w1_rus, w2=w2_rus)
```

```python
# new decision boundary on original dataset
scatter_plot(X, y, plot_boundary=True, b=b_rus, w1=w1_rus, w2=w2_rus)
```
